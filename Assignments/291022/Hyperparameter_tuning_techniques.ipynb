# Model Selection, Preprocessing, performace metrics 

```python
X = dataset['data']
```

```python
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= .30, random_state=1)
```



---



```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_transformed = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_r = LogisticRegression( random_state=42)
```

```python
log_r.fit(X_train, y_train)
```

```python
y_pred = log_r.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
# OR
print("Accuracy:",log_r.score(X_test, y_test))
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
import warnings

warnings.filterwarnings('ignore')
```

```python
log_r_2 = LogisticRegression(random_state=42)
```

```python
import numpy as np
```

```python
parameters = {
    'penalty' : ['l1','l2'], 
    'C'       : np.logspace(-3,3,7),
    'solver'  : ['newton-cg', 'lbfgs', 'liblinear'],
}
```

```python
g_search = GridSearchCV(estimator = log_r_2, param_grid = parameters)
```

```python
g_search.fit(X_train, y_train)
```

```python
print(g_search.best_params_)
print(g_search.best_score_)
```



---



```python
log_r_3 = LogisticRegression(C = 0.1, 
                            penalty = 'l2', 
                            solver = 'liblinear')
log_r_3.fit(X_train,y_train)
y_pred = log_r_3.predict(X_test)
print("Accuracy:", log_r_3.score(X_test, y_test))
```
